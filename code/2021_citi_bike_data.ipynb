```python
import pandas as pd
from pathlib import Path
import datetime as dt
from pyarrow import orc
```

```python
jan_2021 = pd.read_csv(Path('../Resources/2021/jan_2021.csv'), parse_dates=True, infer_datetime_format=True, low_memory=False)
feb_2021 = pd.read_csv(Path('../Resources/2021/feb_2021.csv'), parse_dates=True, infer_datetime_format=True, low_memory=False)
mar_2021 = pd.read_csv(Path('../Resources/2021/mar_2021.csv'), parse_dates=True, infer_datetime_format=True, low_memory=False)
apr_2021 = pd.read_csv(Path('../Resources/2021/apr_2021.csv'), parse_dates=True, infer_datetime_format=True, low_memory=False)
may_2021 = pd.read_csv(Path('../Resources/2021/may_2021.csv'), parse_dates=True, infer_datetime_format=True, low_memory=False)
june_2021 = pd.read_csv(Path('../Resources/2021/june_2021.csv'), parse_dates=True, infer_datetime_format=True, low_memory=False)
july_2021 = pd.read_csv(Path('../Resources/2021/july_2021.csv'), parse_dates=True, infer_datetime_format=True, low_memory=False)
aug_2021 = pd.read_csv(Path('../Resources/2021/aug_2021.csv'), parse_dates=True, infer_datetime_format=True, low_memory=False)
sept_2021 = pd.read_csv(Path('../Resources/2021/sept_2021.csv'), parse_dates=True, infer_datetime_format=True, low_memory=False)
oct_2021 = pd.read_csv(Path('../Resources/2021/oct_2021.csv'), parse_dates=True, infer_datetime_format=True, low_memory=False)
nov_2021 = pd.read_csv(Path('../Resources/2021/nov_2021.csv'), parse_dates=True, infer_datetime_format=True, low_memory=False)
dec_2021 = pd.read_csv(Path('../Resources/2021/dec_2021.csv'), parse_dates=True, infer_datetime_format=True, low_memory=False)
```

```python
jan_2021.head()
```

```python
jan_2021.rename(columns={'starttime': 'started_at', 'stoptime': 'ended_at', 'start station id': 'start_station_id', 'start station name': 'start_station_name', 'start station latitude': 'start_lat', 
                         'start station longitude': 'start_lng', 'end station id': 'end_station_id', 'end station name': 'end_station_name', 'end station latitude': 'end_lat', 
                         'end station longitude': 'end_lng'}, inplace=True)
```

```python
jan_2021.head()
```

```python
feb_2021.head()
```

```python
df_2021 = pd.concat([jan_2021, feb_2021, mar_2021, apr_2021, may_2021, june_2021, july_2021, aug_2021, sept_2021, oct_2021, nov_2021, dec_2021], ignore_index=True)
```

```python
df_2021.drop(columns=['tripduration', 'bikeid', 'birth year', 'gender', 'ride_id', 'rideable_type' ], inplace=True)
```

```python
df_2021.head()
```

```python
df_2021.memory_usage(deep=True).sum()
```

```python
df_2021.insert(0, 'trip_id', range(1, 1 + len(df_2021)))
```

```python
start_date_2021 = pd.to_datetime(df_2021['started_at'], format = '%Y-%m-%d %H:%M:%S')
stop_date_2021 = pd.to_datetime(df_2021['ended_at'], format = '%Y-%m-%d %H:%M:%S')
df_2021['started_at'] = start_date_2021
df_2021['ended_at'] = stop_date_2021
month_start = df_2021['started_at'].dt.strftime('%b')
month_stop = df_2021['ended_at'].dt.strftime('%b')
df_2021['start_month'] = month_start
df_2021['stop_month'] = month_stop
df_2021['time_usage'] = df_2021['ended_at'] - df_2021['started_at']
df_2021.head()
```

```python
df_2021.to_csv('../Resources/2021/2021_citi_bike_data.csv', index=False)
```

```python
df_2021.to_orc('../Cleaned/2021_citi_bike_data.orc')
```
