```python
import pandas as pd
from pathlib import Path
import datetime as dt
```

```python
#Read in 2020 data
jan_2020 = pd.read_csv(Path('../Resources/2020/jan_2020.csv'), parse_dates=True, infer_datetime_format=True)
feb_2020 = pd.read_csv(Path('../Resources/2020/feb_2020.csv'), parse_dates=True, infer_datetime_format=True)
mar_2020 = pd.read_csv(Path('../Resources/2020/mar_2020.csv'), parse_dates=True, infer_datetime_format=True)
apr_2020 = pd.read_csv(Path('../Resources/2020/apr_2020.csv'), parse_dates=True, infer_datetime_format=True)
may_2020 = pd.read_csv(Path('../Resources/2020/may_2020.csv'), parse_dates=True, infer_datetime_format=True)
june_2020 = pd.read_csv(Path('../Resources/2020/june_2020.csv'), parse_dates=True, infer_datetime_format=True)
july_2020 = pd.read_csv(Path('../Resources/2020/july_2020.csv'), parse_dates=True, infer_datetime_format=True)
aug_2020 = pd.read_csv(Path('../Resources/2020/aug_2020.csv'), parse_dates=True, infer_datetime_format=True)
sept_2020 = pd.read_csv(Path('../Resources/2020/sept_2020.csv'), parse_dates=True, infer_datetime_format=True)
oct_2020 = pd.read_csv(Path('../Resources/2020/oct_2020.csv'), parse_dates=True, infer_datetime_format=True)
nov_2020 = pd.read_csv(Path('../Resources/2020/nov_2020.csv'), parse_dates=True, infer_datetime_format=True)
dec_2020 = pd.read_csv(Path('../Resources/2020/dec_2020.csv'), parse_dates=True, infer_datetime_format=True)
```

```python
#Concatenate 2020 data
df_2020 = pd.concat([jan_2020, feb_2020, mar_2020, apr_2020, may_2020, june_2020, july_2020, aug_2020, sept_2020, oct_2020, nov_2020, dec_2020], ignore_index=True)
```

```python
df_2020.rename(columns={'starttime': 'started_at', 'stoptime': 'ended_at', 'start station id': 'start_station_id', 'start station name': 'start_station_name', 'start station latitude': 'start_lat', 
                         'start station longitude': 'start_lng', 'end station id': 'end_station_id', 'end station name': 'end_station_name', 'end station latitude': 'end_lat', 
                         'end station longitude': 'end_lng'}, inplace=True)
```

```python
df_2020.drop(columns=['tripduration', 'bikeid', 'birth year', 'gender'], inplace=True)
```

```python
df_2020.head()
```

```python
#File size of 2020 data
df_2020.memory_usage(deep=True).sum()
```

```python
df_2020.insert(0, 'trip_id', range(1, 1 + len(df_2020)))
```

```python
start_date_2020 = pd.to_datetime(df_2020['started_at'], format = '%Y-%m-%d %H:%M:%S')
stop_date_2020 = pd.to_datetime(df_2020['ended_at'], format = '%Y-%m-%d %H:%M:%S')
df_2020['started_at'] = start_date_2020
df_2020['ended_at'] = stop_date_2020
month_start = df_2020['started_at'].dt.strftime('%b')
month_stop = df_2020['ended_at'].dt.strftime('%b')
df_2020['start_month'] = month_start
df_2020['stop_month'] = month_stop
df_2020['time_usage'] = df_2020['ended_at'] - df_2020['started_at']
df_2020.head()
```

```python
df_2020.to_csv('../Resources/2020/2020_citi_bike_data.csv', index=False)
```
