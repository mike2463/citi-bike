```python
import pandas as pd
from pathlib import Path
import datetime as dt
```

```python
jan_2022 = pd.read_csv(Path('../Resources/2022/jan_2022.csv'), parse_dates=True, infer_datetime_format=True, low_memory=False)
feb_2022 = pd.read_csv(Path('../Resources/2022/feb_2022.csv'), parse_dates=True, infer_datetime_format=True, low_memory=False)
mar_2022 = pd.read_csv(Path('../Resources/2022/mar_2022.csv'), parse_dates=True, infer_datetime_format=True, low_memory=False)
apr_2022 = pd.read_csv(Path('../Resources/2022/apr_2022.csv'), parse_dates=True, infer_datetime_format=True, low_memory=False)
may_2022 = pd.read_csv(Path('../Resources/2022/may_2022.csv'), parse_dates=True, infer_datetime_format=True, low_memory=False)
june_2022 = pd.read_csv(Path('../Resources/2022/june_2022.csv'), parse_dates=True, infer_datetime_format=True, low_memory=False)
july_2022 = pd.read_csv(Path('../Resources/2022/july_2022.csv'), parse_dates=True, infer_datetime_format=True, low_memory=False)
aug_2022 = pd.read_csv(Path('../Resources/2022/aug_2022.csv'), parse_dates=True, infer_datetime_format=True, low_memory=False)
sept_2022 = pd.read_csv(Path('../Resources/2022/sept_2022.csv'), parse_dates=True, infer_datetime_format=True, low_memory=False)
oct_2022 = pd.read_csv(Path('../Resources/2022/oct_2022.csv'), parse_dates=True, infer_datetime_format=True, low_memory=False)
nov_2022 = pd.read_csv(Path('../Resources/2022/nov_2022.csv'), parse_dates=True, infer_datetime_format=True, low_memory=False)
dec_2022 = pd.read_csv(Path('../Resources/2022/dec_2022.csv'), parse_dates=True, infer_datetime_format=True, low_memory=False)
```

```python
df_2022 = pd.concat([jan_2022, feb_2022, mar_2022, apr_2022, may_2022, june_2022, july_2022, aug_2022, sept_2022, oct_2022, nov_2022, dec_2022], ignore_index=True)
```

```python
df_2022.drop(columns=['ride_id', 'rideable_type'], inplace=True)
```

```python
df_2022.head()
```

```python
df_2022.memory_usage(deep=True).sum()
```

```python
start_date_2022 = pd.to_datetime(df_2022['started_at'], format = '%Y-%m-%d %H:%M:%S')
stop_date_2022 = pd.to_datetime(df_2022['ended_at'], format = '%Y-%m-%d %H:%M:%S')
df_2022['started_at'] = start_date_2022
df_2022['ended_at'] = stop_date_2022
month_start = df_2022['started_at'].dt.strftime('%b')
month_stop = df_2022['ended_at'].dt.strftime('%b')
df_2022['start_month'] = month_start
df_2022['stop_month'] = month_stop
df_2022['time_usage'] = df_2022['ended_at'] - df_2022['started_at']
df_2022.head()
```

```python
df_2022.to_csv('../Resources/2022/2022_citi_bike_data.csv', index=False)
```
