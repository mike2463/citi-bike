```python
import pandas as pd
from pathlib import Path
import datetime as dt
```

```python
#Read in 2019 data
jan_2019 = pd.read_csv(Path('../Resources/2019/jan_2019.csv'), parse_dates=True, infer_datetime_format=True)
feb_2019 = pd.read_csv(Path('../Resources/2019/feb_2019.csv'), parse_dates=True, infer_datetime_format=True)
mar_2019 = pd.read_csv(Path('../Resources/2019/mar_2019.csv'), parse_dates=True, infer_datetime_format=True)
apr_2019 = pd.read_csv(Path('../Resources/2019/apr_2019.csv'), parse_dates=True, infer_datetime_format=True)
may_2019 = pd.read_csv(Path('../Resources/2019/may_2019.csv'), parse_dates=True, infer_datetime_format=True)
june_2019 = pd.read_csv(Path('../Resources/2019/june_2019.csv'), parse_dates=True, infer_datetime_format=True)
july_2019 = pd.read_csv(Path('../Resources/2019/july_2019.csv'), parse_dates=True, infer_datetime_format=True)
aug_2019 = pd.read_csv(Path('../Resources/2019/aug_2019.csv'), parse_dates=True, infer_datetime_format=True)
sept_2019 = pd.read_csv(Path('../Resources/2019/sept_2019.csv'), parse_dates=True, infer_datetime_format=True)
oct_2019 = pd.read_csv(Path('../Resources/2019/oct_2019.csv'), parse_dates=True, infer_datetime_format=True)
nov_2019 = pd.read_csv(Path('../Resources/2019/nov_2019.csv'), parse_dates=True, infer_datetime_format=True)
dec_2019 = pd.read_csv(Path('../Resources/2019/dec_2019.csv'), parse_dates=True, infer_datetime_format=True)
```

```python
jan_2019.head()
```

```python
#Concatenate 2019 data
df_2019 = pd.concat([jan_2019, feb_2019, mar_2019, apr_2019, may_2019, june_2019, july_2019, aug_2019, sept_2019, oct_2019, nov_2019, dec_2019], ignore_index=True)
```

```python
df_2019.rename(columns={'starttime': 'started_at', 'stoptime': 'ended_at', 'start station id': 'start_station_id', 'start station name': 'start_station_name', 'start station latitude': 'start_lat', 
                         'start station longitude': 'start_lng', 'end station id': 'end_station_id', 'end station name': 'end_station_name', 'end station latitude': 'end_lat', 
                         'end station longitude': 'end_lng'}, inplace=True)
```

```python
df_2019.drop(columns=['tripduration', 'bikeid', 'birth year', 'gender'], inplace=True)
```

```python
start_date_2019 = pd.to_datetime(df_2019['started_at'], format = '%Y-%m-%d %H:%M:%S')
stop_date_2019 = pd.to_datetime(df_2019['ended_at'], format = '%Y-%m-%d %H:%M:%S')
start_date_2019.head()
```

```python
df_2019['started_at'] = start_date_2019
df_2019['ended_at'] = stop_date_2019
df_2019.head()
```

```python
month_start = df_2019['started_at'].dt.strftime('%b')
month_stop = df_2019['ended_at'].dt.strftime('%b')
```

```python
df_2019['start_month'] = month_start
df_2019['stop_month'] = month_stop
df_2019.head()
```

```python
df_2019['time_usage'] = df_2019['ended_at'] - df_2019['started_at']
```

```python
df_2019.insert(0, 'trip_id', range(1, 1 + len(df_2019)))
```

```python
df_2019.to_csv(Path('../Resources/2019/2019_citi_bike_data.csv'))
```
